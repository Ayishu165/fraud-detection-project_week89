```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load cleaned dataframes 
fraud_data = pd.read_csv("../data/cleaned_fraud_data.csv")
credit_data = pd.read_csv("../data/cleaned_credit_data.csv")
ip_data = pd.read_csv("../data/cleaned_ip_data.csv")
```

```python
#Bivariate Analysis   Goal:Understand how the features relate to fraud
#A. Purchase Value vs Class
plt.figure(figsize=(6,4))
sns.boxplot(data=fraud_data, x='class', y='purchase_value')
plt.title('Purchase Value by Fraud Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Purchase Value')
plt.show()
```

```python
#Bivariate Analysis   Goal:Understand how the features relate to fraud
#B. Age vs Class
plt.figure(figsize=(6,4))
sns.boxplot(data=fraud_data, x='class', y='age')
plt.title('User Age by Fraud Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
#Bivariate Analysis   Goal:Understand how the features relate to fraud
#C. Browser vs Class (Fraud Count by Browser)
plt.figure(figsize=(10,4))
sns.countplot(data=fraud_data, x='browser', hue='class')
plt.title('Fraud vs Non-Fraud by Browser')
plt.xticks(rotation=45)
plt.xlabel('Browser')
plt.ylabel('Count')
plt.legend(title='Class')
plt.show()
```

```python
#Bivariate Analysis   Goal:Understand how the features relate to fraud
#D. Source vs Class
plt.figure(figsize=(6,4))
sns.countplot(data=fraud_data, x='source', hue='class')
plt.title('Fraud by User Source')
plt.xlabel('Source')
plt.ylabel('Count')
plt.legend(title='Class')
plt.show()
```
